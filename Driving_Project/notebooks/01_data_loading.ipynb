```python
# %% 
"""
📌 Notebook: 01_data_loading.ipynb

🎯 Purpose:
This notebook loads raw sensor CSV files from multiple folders,
assigns labels based on driving behavior (1–5 stars),
adds trip identifiers, and merges everything into a single dataset.

Why important?
- Prepares raw data for ML pipeline
- Ensures each trip is treated as an independent sequence
"""

import pandas as pd
import os

print("✅ Libraries loaded successfully.")
```

```python
# %% 
# 📁 Path to dataset folder
dataset_path = "../dataset"

# 🏷️ Mapping folder names to numeric labels
# 1 = Very Rash Driving, 5 = Excellent Driving
ratings = {
    "1 star": 1,
    "2 star": 2,
    "3 star": 3,
    "4 star": 4,
    "5 star": 5
}

print(f"📂 Dataset path set to: {dataset_path}")
```

```python
# %%
"""
🔄 Step: Load all CSV files and assign metadata

We iterate through each rating folder and:
1. Read each CSV file
2. Assign a 'rating' label
3. Assign a unique 'trip_id'
"""

dataframes = []
trip_id_counter = 1  # Unique identifier for each trip/session

# Loop through each rating folder
for folder_name, label in ratings.items():
    folder_path = os.path.join(dataset_path, folder_name)
    
    # Check if folder exists
    if not os.path.exists(folder_path):
        print(f"⚠️ Warning: Folder '{folder_name}' not found. Skipping...")
        continue
    
    # Get all CSV files
    files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
    print(f"📊 Processing {len(files)} files in '{folder_name}'...")
    
    for file in files:
        file_path = os.path.join(folder_path, file)
        
        try:
            # Load CSV
            df = pd.read_csv(file_path)
            
            # 🏷️ Add label (target variable)
            df["rating"] = label
            
            # 🆔 Add trip ID (important for sequence models)
            df["trip_id"] = trip_id_counter
            
            # Append to list
            dataframes.append(df)
            
            trip_id_counter += 1
        
        except Exception as e:
            print(f"❌ Error reading {file}: {e}")

print(f"\n✅ Total unique trips processed: {trip_id_counter - 1}")
```

```python
# %%
"""
📦 Step: Merge all data into a single DataFrame
"""

# Combine all dataframes
dataset = pd.concat(dataframes, ignore_index=True)

# Basic checks
print("📐 Dataset shape:", dataset.shape)

print("\n🔍 First 5 rows:")
display(dataset.head())

print("\n📊 Samples per rating:")
print(dataset['rating'].value_counts().sort_index())
```

```python
# %%
"""
💾 Step: Save merged dataset
"""

output_folder = "../processed_data"

# Create folder if not exists
os.makedirs(output_folder, exist_ok=True)

output_path = os.path.join(output_folder, "merged_dataset.csv")

# Save file
dataset.to_csv(output_path, index=False)

print(f"✅ Success! Merged dataset saved at: {output_path}")
```
